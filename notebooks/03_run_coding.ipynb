## Review Coded Segments

```python
from src.agents.analysis_agent import load_coded_segments

for pid in test_pids:
    coded = load_coded_segments(pid)
    if coded:
        segments = coded.get('coded_segments', []) if isinstance(coded, dict) else coded
        print(f"\n{pid}: {len(segments)} coded segments")
        
        # Show high-quality quotes
        high_q = [s for s in segments if isinstance(s, dict) and s.get('quote_quality', {}).get('overall', 0) >= 4]
        print(f"  High quality quotes: {len(high_q)}")
        for s in high_q[:3]:
            print(f'  ★ "{s.get("text", "")[:120]}..." (score: {s["quote_quality"]["overall"]})')
```
