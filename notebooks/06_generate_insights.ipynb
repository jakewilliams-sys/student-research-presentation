## Review Insights

```python
from src.agents.insight_agent import load_insights

insights = load_insights()
if insights:
    print(f"\n{'='*60}")
    print("EXECUTIVE SUMMARY")
    print(f"{'='*60}")
    print(insights.get('executive_summary', 'Not yet generated'))
    
    print(f"\n\n{'='*60}")
    print(f"INSIGHTS ({len(insights.get('insights', []))})")
    print(f"{'='*60}")
    for ins in insights.get('insights', []):
        if isinstance(ins, dict):
            print(f"\n{ins.get('insight_id', '')}: {ins.get('insight', '')}")
            print(f"  RO: {ins.get('research_objective', '')}")
            print(f"  Confidence: {ins.get('confidence', '')}")
            for r in ins.get('recommendations', []):
                if isinstance(r, dict):
                    print(f"  → [{r.get('type', '')}] {r.get('recommendation', '')}")
```
