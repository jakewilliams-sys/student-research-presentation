```python
import sys, json, logging
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s: %(message)s')

from src.agents.deep_dive_agent import DeepDiveAgent
from src.pipeline.orchestrator import PipelineOrchestrator
```

## Configure Deep Dive

Edit the parameters below to focus the deep dive.

```python
FOCUS_AREA = 'Post-graduation habit evolution'
PARTICIPANTS = ['P024', 'P025', 'P026', 'P027', 'P028']  # Early graduates
QUESTIONS = [
    'How do cooking habits change after graduation?',
    'Does the emotional relationship with delivery shift?',
    'What triggers reactivation vs lapsing?',
]
HYPOTHESIS = {
    'current': 'Graduates order less due to better cooking skills',
    'alternative': 'Graduates order differently (e.g. more solo, less social) but not necessarily less',
}
```

```python
orch = PipelineOrchestrator()

context = {
    'dive_id': 'DD_001',
    'focus_area': FOCUS_AREA,
    'participants_to_revisit': PARTICIPANTS,
    'specific_questions': QUESTIONS,
    'alternative_hypothesis': HYPOTHESIS,
    'transcripts': {},
}

# Load transcripts for specified participants
for pid in PARTICIPANTS:
    t = orch._load_transcript(pid, orch.participant_map.get(pid, {}))
    if t:
        context['transcripts'][pid] = t

print(f'Loaded transcripts for {len(context["transcripts"])} / {len(PARTICIPANTS)} participants')
```

```python
agent = DeepDiveAgent()
output = agent.run('DD_001', context)

print(f'Success: {output.success}')
if output.data:
    print(json.dumps(output.data, indent=2, default=str)[:3000])
```
