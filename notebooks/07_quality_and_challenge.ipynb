```python
import sys, json, logging
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT))

logging.basicConfig(level=logging.INFO, format='%(asctime)s %(levelname)s: %(message)s')

from src.pipeline.orchestrator import PipelineOrchestrator
```

```python
orch = PipelineOrchestrator()

# Run QA
qa_results = orch.run_agent('quality')
print('QA:', json.dumps(qa_results, indent=2, default=str))

# Run Devil's Advocate
adv_results = orch.run_agent('advocate')
print('\nAdvocate:', json.dumps(adv_results, indent=2, default=str))
```

## Review QA Results

```python
from src.agents.quality_agent import load_qa_results

qa = load_qa_results()
if qa:
    overall = qa.get('overall_assessment', {})
    print(f"Evidence coverage: {overall.get('evidence_coverage_pct', '?')}%")
    print(f"Contradictions: {overall.get('contradiction_count', '?')}")
    print(f"Low confidence findings: {overall.get('low_confidence_count', '?')}")
    
    critical = qa.get('critical_issues', [])
    if critical:
        print(f"\n⚠️ CRITICAL ISSUES ({len(critical)}):")
        for issue in critical:
            print(f"  - {issue}")
```

## Review Devil's Advocate Report

```python
from src.agents.advocate_agent import load_advocate_results

adv = load_advocate_results()
if adv:
    print(f"Findings strengthened: {adv.get('findings_strengthened', '?')}")
    print(f"Findings weakened: {adv.get('findings_weakened', '?')}")
    
    bias = adv.get('bias_assessment', {})
    if bias:
        print(f"\nBias Assessment:")
        for k, v in bias.items():
            print(f"  {k}: {v}")
    
    blind = adv.get('blind_spots', [])
    if blind:
        print(f"\nBlind Spots ({len(blind)}):")
        for b in blind:
            print(f"  - {b}")
```
