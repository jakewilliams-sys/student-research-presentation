## Review Summaries

```python
from src.agents.summary_agent import load_summary

for pid in test_pids:
    s = load_summary(pid)
    if s:
        print(f"\n{'='*60}")
        print(f"SUMMARY: {pid}")
        print(f"{'='*60}")
        print(f"\n{s.get('summary', 'No summary')}")
        print(f"\nFood relationship: {s.get('food_relationship', '')}")
        print(f"\nKey moments: {len(s.get('key_moments', []))}")
        for km in s.get('key_moments', []):
            print(f"  - [{km.get('timestamp', '')}] {km.get('description', '')}")
        print(f"\nHypotheses: {s.get('initial_hypotheses', [])}")
    else:
        print(f"No summary found for {pid}")
```
